### 3D json 확인

```python
import pandas as pd
import json

# 1. 파일 업로드
from google.colab import files
uploaded = files.upload()

# 2. 업로드된 파일 이름 확인 (파일 이름이 'data.json'이라고 가정)
file_name = '3108M_BD_A.json'

# 3. JSON 파일 읽기
with open(file_name, 'r', encoding='utf-8') as file:
    data = json.load(file)

# 4. JSON 데이터 확인
print(json.dumps(data, indent=2, ensure_ascii=False))

# 5. keypoints 데이터를 데이터 프레임으로 변환
keypoints = data['keypoints']
df_keypoints = pd.DataFrame(keypoints)

# 6. keypoints 데이터 프레임 출력
print(df_keypoints)

# 7. 필요에 따라 다른 부분도 데이터프레임으로 변환하여 출력
category = data['category']
annotation = data['annotation']
mesh = data['mesh']
actor = data['actor']
pose = data['pose']
license_info = data['license']
camera = data['camera']
info = data['info']

# 각 데이터를 데이터프레임으로 변환 (필요시)
df_category = pd.DataFrame([category])
df_annotation = pd.DataFrame([annotation])
df_mesh = pd.DataFrame([mesh])
df_actor = pd.DataFrame([actor])
df_pose = pd.DataFrame([pose])
df_license = pd.DataFrame([license_info])
df_camera = pd.DataFrame([camera])
df_info = pd.DataFrame([info])

# 각 데이터프레임 출력
print(df_category)
print(df_annotation)
print(df_mesh)
print(df_actor)
print(df_pose)
print(df_license)
print(df_camera)
print(df_info)
```
